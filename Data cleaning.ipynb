```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/companies.csv")
df
```

1. Handling duplicates

```python
df.drop_duplicates()
```

2. Drop column

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
df
```

```python
df[["City","State"]]=df['Headquarters'].str.split(',',n=1,expand=True)
df
```

```python
df = df.drop(columns=["Headquarters"])
df
```

```python
df['Revenue (USD millions)'] = df['Revenue (USD millions)'].astype(str).str.replace(',', '').astype(int)
df['Revenue growth'] = df['Revenue growth'].astype(str).str.replace('%', '').astype(float)
df['Employees'] = df['Employees'].astype(str).str.replace(',', '').astype(int)
df = df.astype({"Employees": int})
df = df.convert_dtypes()
df

```
